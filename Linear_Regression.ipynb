## Task 3: Multiple Linear Regression Analysis

### Loading data

```python
data_df=df[['cyl','wt','hp','mpg']]
data_df.head()
```

```python
print(data_df.shape)
```

```python
# Checking for null value in each column

print(data_df.isnull().sum())
new_df1=data_df
```

### Splitting the dataset into train and test

```python
Xm, ym = new_df1.iloc[:, :-1], new_df1.iloc[:, [-1]] 
```

```python
Xm_train, Xm_test, ym_train, ym_test = train_test_split(Xm, ym, test_size=0.3,random_state=100)
```

```python
ym_test
```

### Training the model

```python
model_mul = LinearRegression().fit(Xm_train, ym_train)

```

###Evaluating the model performance

```python
print('Intercept: \n', model_mul.intercept_)
print('Coefficients: \n', model_mul.coef_)
```

#model equation

mpg = -0.81* cyl -3.20 *wt -0.013*hp + 37.50

```python
#MSE and RMSE
MSE=mean_squared_error(ym_test, model_mul.predict(Xm_test))
RMSE = mean_squared_error(ym_test, model_mul.predict(Xm_test), squared=False)

#R2 score
r_sq2 = model_mul.score(Xm_train,ym_train)
print('coefficient of determination:', r_sq2)
print('MSE and RMSE is',MSE, RMSE)

#RMSE tells us how the residuals are spreaded
```

 Inferences:
 R2 coefficient tells how well the regression predictions approximate the real data points. An R2 of 1 indicates that the regression predictions perfectly fit the data. SInce the value here is 0.87, the model is a good model

 MSE and RMSE
  MSE measures the variance of the residuals.
  RMSE measures the standard deviation of residuals. The lower the value of MSE and RMSE, higher the accuracy of the model
